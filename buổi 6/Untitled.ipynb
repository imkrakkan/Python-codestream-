```python
a=6
b=7
c=8
avg=(a+b+c)/3
#điểm trung bình nằm trong mức giỏi 
if avg >= 8 : 
    print(True)
#điểm trung bình nằm trong mức khá 
elif 6 <= avg < 8:
    print(True)

#điểm trung bình nằm trong mức trung bình không
elif avg < 6 : 
    print(True)
else:
    print(False)
```
