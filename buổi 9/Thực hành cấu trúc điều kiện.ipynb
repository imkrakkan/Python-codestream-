### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
gia_ga_ran = 35000
gia_hamburger = 50000
gia_pesi = 10000

sl_ga_ran_khach_mua = int(input("Số lượng Gà rán khách mua: "))
sl_hamburger_khach_mua = int(input("Số lượng Hamburger khách mua: "))
sl_pesi_khach_mua = int(input("Số lượng Pesi khách mua: "))

tong_tien_don_hang = sl_ga_ran_khach_mua * gia_ga_ran + sl_hamburger_khach_mua * gia_hamburger + sl_pesi_khach_mua * gia_pesi

if tong_tien_don_hang > 4000000:
    giam_gia = tong_tien_don_hang * 0.15
    print(f"Tổng tiền đơn hàng là: {tong_tien_don_hang}")
    print(f"Tổng tiền được giảm 15% là: {giam_gia}")
elif tong_tien_don_hang > 2000000:
    giam_gia = tong_tien_don_hang * 0.1
    print(f"Tổng tiền đơn hàng là: {tong_tien_don_hang}")
    print(f"Tổng tiền được giảm 10% là: {giam_gia}")
else:
    print(f"Tổng tiền đơn hàng là: {tong_tien_don_hang}")

print("Cảm ơn quý khách đã mua")
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

````python
chieu_cao = int(input("nhập chiều cao:"))
can_nang = int(input("nhập cân nặng:"))
BMI = chieu_cao/ pow(can_nang,2)
if BMI < 16: print("Gầy cấp độ III")
elif 16 <= BMI < 17 : print("Gầy cấp độ II")
elif 17<= BMI < 18.5 : print("Gầy cấp độ I")
elif 18.5 <= BMI < 25 : print("Bình thường")
elif 25 <= BMI < 30 : print("Thừa cân")
elif 30 <= BMI < 35 : print("Béo phì cấp độ I")
elif 35 <= BMI < 40 : print("Béo phì cấp độ II")
else : print("Béo phì cấp độ III")
                            
````
